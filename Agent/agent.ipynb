#### MRKL论文 
https://arxiv.org/abs/2205.00445  
让大模型可以调用外部工具。增加了一个路由模块，可以将自然语言转化为工具输入，通过工具来执行专业任务。  
这个设计比较完备，工具不止是一些调用接口，还可以是一些专有模型。如果没有工具命中则传给大模型直接回答。  
我理解，这篇论文的意义是，给大模型装上了手脚，训了一个模型来将大模型的指令翻译成手脚能用的指令，从而可以指挥手脚去做任务。  

### ReAct论文
https://arxiv.org/abs/2210.03629  
上一篇论文中MRKL任务处理是线性的，只是把任务分解、执行，之后就组织答案了。  
这篇论文ReAct是仿照人的思考、行动、观察模式，对一个问题进行拆解。  
在LLM中，通过提示词引导LLM给出思考、行动的步骤，然后行动，行动可以理解为上文中的工具调用。获取结果。然后再将结果反馈给LLM，再指导下一轮的行动。  
我理解，这篇论文的意义是，虽然MRKL给LLM装上了手脚，但LLM没能更好的利用手脚。ReAct通过提示词让LLM可以思考、规划行动并读取反馈，形成了一个闭环。LLM不仅能行动，还能自己思考规划了。  

### 实现  
opencode:https://opencode.ai/  

总流程相对比较固定，说白了就是访问LLM，拿到回答，解析其中的tool_call，执行tool_call，返回tool_result给LLM。以此循环。  
是有一些手段可以对行为进行控制，包括权限控制、命令组装、Skills等等，但我看主要还是提示词+模型效果，决定了智能程度。这些也都是提供给用户的接口，无需用户去定制化开发Agent。
