想到一个问题：既然agent可以把任何机器码反编译成源码、自然语言给人看，那是不是相当于所有的程序的源码都开源了呢？  
不是的。源码本身是有结构的，比如各种对象的抽象、各种层的封装等。这些是一个项目既有各种功能又可以长期迭代的关键。编译成机器码之后，这些结构信息都丢失了，且不能复原。就好像你把一个三维空间投影到了一维一样。你只知道一维投影什么样，用处不大。

stream流式  
非流式是发送请求，模型一直等输出结束，将整个输出返回。流式是，模型输出一部分就返回一部分，具体多少token，和模型的配置有关。客户端收到模型返回的部分结果后，依次append就可以了。流式通过HTTP的Chunked Transfer Encoding实现，返回的时候可以分块返回，通过TCP保序。其中的内容通过SSE格式传输。SSE就类似rpc里的protobuf，一种http之上的协议。
