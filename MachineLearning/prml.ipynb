#### 均方误差(RMS)和平方误差和比，有什么区别？  
平方误差和：  
$E(\mathbf{w}) = \sum_{n=1}^N\{y(x_n, \mathbf{w}) - t_n\}^2$  
均方误差：  
$E_{RMS}(\mathbf{w}) = \sqrt{2E(\mathbf{w})/N}$  
$N$可以消去数据集大小的影响，让不同数据集之间可比。  
$\sqrt{}$让误差落回和目标$t$一样的尺度，可以和$t$来比较看差异。
