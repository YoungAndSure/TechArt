为什么说bootstrap法是频率派的方法？bootstrap是说，把样本X有放回的随机抽取N次，得到N组样本$X_n$，用这N组样本计算指标（如loss等），得出指标的误差/置信区间等。对样本重新随机抽取N次，其实就是频率派的思想：对事件重复多次，某事件发生的频率就是概率。只不过理论上，需要真实的让事件发生来产生样本，比如硬币抛1000次，而现实中只能拿到已经抛了1000次的样本，很难再用相同情况重新抛1000次。所以才用已收集的样本，随机抽样100个样本来估计，就是模拟重新抛了100次硬币。这个过程重复N次，根据估计结果，得出误差/置信区间等，思想还是重复N次，通过频率算概率。  
贝叶斯不同的是，已经有了假设先验分布，直接把所有样本输入进去，就得到后验分布，因为是个分布，所以天然就可以得到误差/置信区间等，不需要重复。
